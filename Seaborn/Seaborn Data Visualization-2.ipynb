# 6. Multiplots / MultiGrid plot

<h3>(i)FacetGrid</h3>

```python
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.scatterplot,'sex','total_bill')
g.add_legend()
```

<h3>(ii)piarplot</h3>

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

<h3>(iii)PiarGrid</h3>

```python
g=sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
# map_diag, map_offdiag
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag, map_upper, map_lower
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# vars
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

<h3>(iv)jointPlot vs jointGrid</h3>

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
#JointGrid
g=sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.boxplot)
```

# 1. Relational plot

<h3>(i)Scatter plot</h3>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
#basically sctterplot is a axis level function 
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
#relplot is a figure level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
 sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

# 2. Distribution plot

<h3>(i) histplot</h3>

```python
#axis level function
sns.histplot(data=tips,x='total_bill')
```

```python
#figure level function
sns.displot(data=tips,kind='hist',x='total_bill')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
# histplot for the categorical column
sns.displot(data=tips,kind='hist',x='day')
```

```python
#hue parameter
sns.displot(kind='hist',data=tips,x='tip',hue='sex')
```

```python
#element = step
sns.displot(data=tips,kind='hist',x='tip',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.displot(data=titanic,kind='hist',x='age',hue='sex',element='step')
```

```python
#feceting using col and row
sns.displot(data=titanic,kind='hist',x='age',col='sex')
```

```python
sns.displot(data=tips,kind='kde',x="total_bill",hue='sex')
```

# 3.Matrix plot

<h3>(i)Heatmap</h3>

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
temp
```

```python
# annot parameter
plt.figure(figsize=(15,15))
sns.heatmap(temp,annot=True)
```

```python
#linewidht parameter
plt.figure(figsize=(15,15))
sns.heatmap(temp,annot=True,linewidth=0.5)
```

```python
#cmap parameter
plt.figure(figsize=(15,15))
sns.heatmap(temp,annot=True,linewidth=0.5,cmap='viridis')
```

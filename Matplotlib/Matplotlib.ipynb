<h3>Annotation</h3>

```python
batter=pd.read_csv(r"D:\Navnit AI,ML\Datasets\matplotlib Dataset\batter.csv")
```

```python
batter
```

```python
df_sample=batter.head(100).sample(25,random_state=5)
```

```python
df_sample
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df_sample['avg'],df_sample['strike_rate'],s=df_sample['runs']/10)

for i in range(df_sample.shape[0]):
    plt.text(df_sample['avg'].values[i],df_sample['strike_rate'].values[i],df_sample['batter'].values[i])
```

<h3>Horizontal and vertical line</h3>

```python
plt.figure(figsize=(15,7))
plt.scatter(df_sample['avg'],df_sample['strike_rate'],s=df_sample['runs']/10)

plt.axhline(130,color='red')
plt.axvline(30,color='red')
plt.axhline(140,color='green')

for i in range(df_sample.shape[0]):
    plt.text(df_sample['avg'].values[i],df_sample['strike_rate'].values[i],df_sample['batter'].values[i])
```

<h3>Subplots</h3>

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0].set_title("Average vs Strike Rate")
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("Strike Rate")
ax[1].scatter(batter['avg'],batter['runs'])
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")
ax[1].set_title("Average vs Runs")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'],color="blue")
ax[1,1].hist(batter['runs'])
```

<h3 style={color:red}> Another way to plot multiple graphs</h3>

```python
fig=plt.figure()

ax0=fig.add_subplot(2,2,1)
ax0.scatter(batter['avg'],batter['strike_rate'])

ax1=fig.add_subplot(2,2,2)
ax1.hist(batter['runs'],color='red')

ax2=fig.add_subplot(2,2,3)
ax2.hist(batter['avg'])
```

<h3>3D surface plot</h3>

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x.shape
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx.shape
```

```python
z=xx**2+yy**2
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
a=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(a)
```

<h3> contour plot</h3>

```python
fig=plt.figure()
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

<h3>Opening a csv file from an URL</h3>

```python
import requests
from io import StringIO

url='https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv'

headers={"User_Agent":"Mozilla/5.0(macintosh; Intel mac os x 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url,headers=headers)
data=StringIO(req.text)
pd.read_csv(data)
```
